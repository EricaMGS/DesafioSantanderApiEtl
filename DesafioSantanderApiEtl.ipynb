## Criação de um DataFrame nome tendencias e consulta 5 valores mais recorrentes

```python
import pandas as pd

tendencias = pd.read_csv("shopping_trends_updated.csv")
tendencias.head()
```

## Consulta de todos os valores da coluna: Item Purchased (item adquirido)

```python
 tendencias['Item Purchased']
```

##Consulta frequência de  compra de cada item comprado, quantidade de vezes que foi comprado

```python
tendencias['Item Purchased'].value_counts()
```

## Consulta nome itens da coluna Item Purchased

```python
itens = tendencias['Item Purchased'].unique()
print(itens)
```

## Consulta média frequência de compra dos itens

```python
quantidade = tendencias['Item Purchased'].value_counts()
media = quantidade.mean()
print(f'Média = {media:.0f}')
```

## Gráfico quantidade comprada cada item

```python
quantidade.plot()
```

## Média Idade Consumidor

```python
media_idade_consumidor = tendencias.Age.mean()
print(f'Idade média do consumidor: {media_idade_consumidor:.0f}')
```

## Consulta quantidade consumidores por gênero

```python
tendencias['Gender'].value_counts()
```

## Gráfico quantidade consumidores por gênero

```python
quantidade_comprada_por_genero = tendencias['Gender'].value_counts()
quantidade_comprada_por_genero.plot(kind='bar')
```
